```python
'''
ATM program brought to you courtesy of code-with-Gideon Academy.
Enjoy the program!
Remember you have only 3 attempts to enter your pin correctly
'''
# importing useful libraries
import time as t

# welcome screen
print('@'*20)
print('--- Welcome to Giddy Bank PLC --- \n')
print('@'*20)
user_name = input('Please Enter your user_name: ')

# creating the ATM program engine
def atm_pro():

    # request pin function
    def request_pin():
        while True:
            # Getting user 4 digit pin
            secret_pin = input('Enter your 4 digit PIN: ')

            # checks for data type of pin
            if secret_pin.isdecimal():
                return int(secret_pin)
            print('\nPIN incorrect! \nPlease enter your 4 digit PIN')

    # new transaction
    def new_trans():
      # To perform another transaction
      new_trans = input('Do you wish to perform another transaction? (Yes/No): ')
      while new_trans.lower() == 'yes':
        atm_pro()
        break
      else:
        print('Exiting in progress...')
        t.sleep(2)
        print('You have been successfully logged out \n')
        print('--- Thank you for choosing Giddy Bank PLC! ---')


    # creating number of attempts (3)
    for i in range(3):
        # creating a user instance
        user_pin = 1234
        user_balance = 50000

        # checking if secret pin matches
        secret_pin = request_pin()
        if (secret_pin == user_pin):
            # upon successful login
            print('@'*20)
            print(f'Login successful!\nWelcome, {user_name}. \nSelect the transaction you wish to perform.')
            print('@'*20)

            # Transaction options
            print('')
            print('Enter 0: Logout and Exit \n')
            print('Enter 1: Balance Enquiry \n')
            print('Enter 2: Withdraw Cash \n')
            print('Enter 3: Quick Deposit \n')
            print('Enter 4: Cash Transfer \n')
            print('Enter 5: Change PIN \n')
            print('Enter 6: Return Card \n')
            print('@'*20)

            # Getting user transaction option
            option = int(input('Please enter your choice transaction number to proceed: '))
            print('')

            # Logout option
            if (option == 0):
                print('Exiting in progress...')
                t.sleep(2)
                print('You have been successfully logged out \n')
                print('--- Thank you for choosing Giddy Bank PLC! ---')
                break

            # Other bank transaction options
            elif option in (1, 2, 3, 4, 5, 6):
                # Balance Enquiry
                if (option == 1):
                    print('loading account balance...')
                    t.sleep(1.5)
                    print(f'Hey {user_name}.\nYour account balance is ${user_balance} \n')

                    # To perform another transaction
                    new_trans = new_trans()
                    break

                # Withdraw cash
                elif (option == 2):
                    print('Welcome to Giddy Bank Cash Withdraw \n')
                    t.sleep(1.5)
                    print('Enter 1: $5000 \n')
                    print('Enter 2: $10000 \n')
                    print('Enter 3: $15000 \n')
                    print('Enter 4: $20000 \n')
                    print('Enter 5: other amounts greater than $25000 \n')

                    choice = int(input('Enter your withrawal option: '))

                    if (choice == 1 and user_balance > 5000):
                        user_balance = user_balance - 5000
                        # user_balance -= 5000
                        print('--- Please take your cash --- \n')
                        print(f'--- Your current account balance is ${user_balance} ---')

                        # To perform another transaction
                        new_trans = new_trans()
                        break

                    elif (choice == 2 and user_balance > 10000):
                        user_balance = user_balance - 10000
                        print('--- Please take your cash --- \n')
                        print(f'--- Your current account balance is ${user_balance} ---')

                        # To perform another transaction
                        new_trans = new_trans()
                        break

                    elif (choice == 3 and user_balance > 15000):
                        user_balance = user_balance - 15000
                        print('--- Please take your cash --- \n')
                        print(f'--- Your current account balance is ${user_balance} ---')

                        # To perform another transaction
                        new_trans = new_trans()
                        break

                    elif (choice == 4 and user_balance > 20000):
                        user_balance = user_balance - 20000
                        print('--- Please take your cash --- \n')
                        print(f'--- Your current account balance is ${user_balance} ---')

                        # To perform another transaction
                        new_trans = new_trans()
                        break

                    elif (choice == 5 and user_balance > 25000):
                        withdraw_amount = int(input('Enter the amount to withdraw: '))
                        if (user_balance > withdraw_amount and withdraw_amount % 500 == 0 ):
                            user_balance = user_balance - withdraw_amount
                            print('--- Please take your cash --- \n')
                            print(f'Your current account balance is ${user_balance}')

                            # To perform another transaction
                            new_trans = new_trans()
                            break
                        else:
                            print('Please enter an amount that is in multiples of $500')
                            # To perform another transaction
                            new_trans = new_trans()
                            break
                    else:
                        print('Please enter a valid transaction option')
                    break

                # Quick Deposit
                elif (option == 3):
                    print('--- Welcome to Giddy Bank Quick deposit --- \n')
                    t.sleep(1.5)
                    deposit_amount = int(input('Enter amount to deposit: '))
                    user_balance = deposit_amount + user_balance
                    print(f'--- You have successfully deposited ${deposit_amount} ---\n')
                    print(f'--- Your current account balance is ${user_balance} ---')

                    # To perform another transaction
                    new_trans = new_trans()
                    break

                # Cash Transfer
                elif (option == 4):
                    print('--- Welcome to Giddy Bank Quick Cash transfer --- \n')
                    t.sleep(1.5)
                    # Receiver's account details
                    bank_name = input('Enter the receiver bank name: ')
                    acc_num = int(input('Enter the receiver account number: '))
                    acc_name = input('Enter the receiver account name: ')

                    trans_amount = int(input('Enter the amount to transfer: '))
                    user_balance = user_balance - trans_amount
                    t.sleep(1.5)
                    print(f'--- You have successfully transferred ${trans_amount} to: \nAccount number: {acc_num} \n Account name: {acc_name} \n Bank name: {bank_name} ---')

                    # To perform another transaction
                    new_trans = new_trans()
                    break

                # Change PIN
                elif (option == 5):
                    print('Loading change PIN...')
                    t.sleep(1.5)
                    new_pin = int(input('Enter your new 4-digit PIN: '))
                    print(f'changing PIN to {new_pin}')
                    confirm = input('Confirm? Y/N: ')
                    if confirm in ('Y', 'y'):
                        user_pin = new_pin
                        print('--- PIN changed successfully! ---\n')
                    else:
                        print('Cancelling PIN change...')
                        t.sleep(1)
                        print('Process Cancelled! \n')

                    # To perform another transaction
                    new_trans = new_trans()
                    break

                # Return card
                elif (option == 6):
                    print('Loading card return...')
                    t.sleep(1.5)
                    print('Returning your ATM card')
                    confirm = input('Confirm? Y/N: ')
                    if confirm.lower() == 'y':
                        print('--- Please take your ATM card --- \n')
                        print('--- Card returned successfully! ---')
                    else:
                        print('Cancelling process...')
                        t.sleep(1)
                        print('Proccessed cancelled!')

                else:
                    print('Please enter a valid transaction number')
            else:
                print('Kindly enter a valid transaction option!')
        else:
            print('\nPin incorrect!\nYou have {} attempts left to enter your secret pin correctly!'.format(2 - i))
    else:
        print('\nDo you wish to create an account with Giddy Bank Plc?')

atm_pro()
```
